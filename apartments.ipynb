```python
apartments.dtypes
apartments['zip_code'] = apartments['zip_code'].astype(int)
apartments['google_star'] = apartments['google_star'].astype(float)
def handle_price(price):
    if isinstance(price, str):
        if '-' in price:
            low, high = price.replace('$', '').replace(',', '').replace('+', '').split('-')
            return [float(low), float(high)]
        else:
            return [float(price.replace('$', '').replace(',', '').replace('+', ''))]
    else:
        return price

apartments['price'] = apartments['price'].apply(handle_price)

apartments = apartments.sort_values(by='price', ascending=True)


# Load your data
def handle_price2(price):
    if isinstance(price, list):
        return price[0]
    else:
        return price

apartments['price'] = apartments['price'].apply(handle_price2)
apartments['price'] = apartments['price'].astype(int)
apartments.dtypes
```
